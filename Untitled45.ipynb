```python
import random
import numpy as np
import pandas as pd
from pulp import LpMaximize, LpProblem, LpVariable, lpSum

# -----------------------------
# 1. Generate Synthetic SMB Customer Data
# -----------------------------

NUM_CUSTOMERS = 1000    # Number of SMBs
MAX_VISITS_PER_CUSTOMER = 5  # We cap visits to at most 5 in this example
TOTAL_VISITS_BUDGET = 2000   # Total visits you can make

random.seed(42)  # For reproducibility

# Each customer has:
# - base_potential: how likely they are to convert at baseline
# - sensitivity_to_visits: how strongly they respond to additional visits
# - diminishing returns after a certain point
customer_data = []
for i in range(NUM_CUSTOMERS):
    base_potential = random.uniform(0.05, 0.25)          # base conversion prob (just a random range)
    sensitivity = random.uniform(0.10, 0.30)             # how strongly additional visits help
    diminishing_factor = random.uniform(0.7, 0.95)       # how quickly additional visits' effect tapers
    customer_data.append({
        'customer_id': i,
        'base_potential': base_potential,
        'sensitivity': sensitivity,
        'diminishing_factor': diminishing_factor
    })

df_customers = pd.DataFrame(customer_data)

# For illustration, define a function that, given (v) visits, returns expected conversion probability.
# We assume each additional visit helps, but by a diminishing factor:
def expected_conversion_prob(base, sensitivity, diminishing_factor, visits):
    """
    A toy function that models a conversion probability
    given a certain number of visits.
    """
    prob = base
    for v in range(visits):
        # Each visit adds some fraction of 'sensitivity',
        # but that fraction is multiplied each time by the diminishing_factor.
        prob += sensitivity * (diminishing_factor ** v)
    # Cap max probability at 1.0
    return min(prob, 1.0)

# ---------------------------------------------
# 2. Constrained Optimization to Assign Visits
# ---------------------------------------------

# We want to decide: how many visits for each customer (0 to MAX_VISITS_PER_CUSTOMER)
# to maximize sum of expected conversions, subject to a total visits budget.

model = LpProblem(name="Dynamic_Targeting", sense=LpMaximize)

# Decision variables: integer number of visits per customer
visits_vars = {
    i: LpVariable(name=f"visits_{i}", lowBound=0, upBound=MAX_VISITS_PER_CUSTOMER, cat='Integer')
    for i in df_customers.index
}

# Objective: Maximize sum of expected conversions
# We need the expected conversion for each (customer, # visits), but # visits is a variable.
# Because that relationship is nonlinear, we might approximate or handle it in different ways.
# For simplicity, let's precompute the expected conversions for each possible number of visits,
# then we'll use an approach called "piecewise approximation" or simply do a direct table lookup.

# But direct table lookup still poses a non-linear expression.
# To keep it simpler in code, we can do a brute force approach or do an integer linear approximation.
# However, for demonstration, let's do a simpler approach:
# We'll just compute the expected conversion for each integer possibility from 0..MAX_VISITS,
# and then multiply by a "binary assignment" approach or solve with an external solver.
# 
# But let's do a simpler method: 
# We'll keep the objective as a sum(prob * (some binary var)) approach.
# This can get tricky. For demonstration let's do a small hack:
#   We'll treat the expected conversion for each integer # of visits as a piecewise constant function
#   We can do a single variable approach but it's technically not purely linear. 
# 
# For a blog example, let's do a simpler "greedy" approach in code and skip the full MILP complexity.

# --- ALTERNATIVE: We'll do a simple *greedy heuristic* for demonstration. ---
#  1. For each possible (customer, additional visit), compute the marginal gain in expected conversions.
#  2. Pick the best marginal gain until we run out of visits.

# Precompute expected conversion probabilities for each (customer, 0..MAX_VISITS_PER_CUSTOMER).
prob_matrix = {}
for i, row in df_customers.iterrows():
    prob_matrix[i] = []
    for v in range(MAX_VISITS_PER_CUSTOMER + 1):
        prob_matrix[i].append(
            expected_conversion_prob(
                row['base_potential'],
                row['sensitivity'],
                row['diminishing_factor'],
                v
            )
        )

# Now let's do a greedy approach:
# Start everyone at 0 visits, then repeatedly assign 1 more visit to the customer who yields
# the highest marginal gain in expected conversions, until we run out of visits.

df_customers['assigned_visits'] = 0
df_customers['expected_conversion'] = prob_matrix[df_customers.index[0]][0]  # placeholder init

assigned_visits = [0]*NUM_CUSTOMERS
remaining_visits = TOTAL_VISITS_BUDGET

while remaining_visits > 0:
    best_customer = None
    best_gain = 0.0

    for i in range(NUM_CUSTOMERS):
        current_visits = assigned_visits[i]
        if current_visits < MAX_VISITS_PER_CUSTOMER:
            current_prob = prob_matrix[i][current_visits]
            next_prob = prob_matrix[i][current_visits + 1]
            marginal_gain = next_prob - current_prob
            if marginal_gain > best_gain:
                best_gain = marginal_gain
                best_customer = i
    
    if best_customer is None:
        # That means no improvement found or all customers maxed out
        break
    else:
        assigned_visits[best_customer] += 1
        remaining_visits -= 1

df_customers['assigned_visits'] = assigned_visits
df_customers['optimal_expected_conversion'] = [
    prob_matrix[i][assigned_visits[i]] for i in range(NUM_CUSTOMERS)
]

# Compute total expected conversions from the "optimal" assignment
optimal_total_conversions = df_customers['optimal_expected_conversion'].sum()

print("Greedy Optimization Approach")
print(f"Total Visits Allocated: {sum(assigned_visits)}")
print(f"Total Expected Conversions (Optimal): {optimal_total_conversions:.2f}\n")

# -------------------------
# 3. Observational Experiment
# -------------------------
# Suppose in real life, the actual number of visits is distributed with some randomness.
# Some customers match the "optimal" number, some get fewer, some get more.

actual_visits = []
for i in range(NUM_CUSTOMERS):
    opt_vis = assigned_visits[i]
    # Simulate that actual visits might be plus or minus 1 from the optimal, within [0, MAX].
    # Weighted to usually match the optimal, but not always.
    possible_visits = []
    # Weighted chances: 50% to get exactly the optimal, 25% for less, 25% for more (if feasible).
    # This is just a random scheme for illustration.
    draw = random.random()
    if draw < 0.5:
        # exactly optimal
        actual_vis = opt_vis
    elif draw < 0.75:
        # one less if possible
        actual_vis = max(opt_vis - 1, 0)
    else:
        # one more if possible
        actual_vis = min(opt_vis + 1, MAX_VISITS_PER_CUSTOMER)

    actual_visits.append(actual_vis)

df_customers['actual_visits'] = actual_visits
df_customers['actual_conversion_prob'] = [
    prob_matrix[i][actual_visits[i]] for i in range(NUM_CUSTOMERS)
]

# Now let's assume a Bernoulli draw for actual conversions (like flipping a biased coin).
df_customers['actual_conversion'] = [
    1 if random.random() < df_customers.loc[i, 'actual_conversion_prob'] else 0
    for i in range(NUM_CUSTOMERS)
]

# Group customers by whether they got "Less", "Equal", or "More" visits than recommended
def compare_visits(row):
    if row['actual_visits'] < row['assigned_visits']:
        return 'Less'
    elif row['actual_visits'] == row['assigned_visits']:
        return 'Equal'
    else:
        return 'More'

df_customers['visit_comparison_group'] = df_customers.apply(compare_visits, axis=1)

# Compute actual conversion rates in each group
grouped = df_customers.groupby('visit_comparison_group')['actual_conversion'].mean()
print("Observational Experiment Results:")
print(grouped)

print("\nSample of final DataFrame:")
print(df_customers.head(10))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Let’s assume df_customers is already defined from the previous code.
# For consistent style
sns.set_theme(style="whitegrid")
```

```python
# Group-level mean of actual conversion (binary 0/1)
group_means = df_customers.groupby('visit_comparison_group')['actual_conversion'].mean().reset_index()

plt.figure(figsize=(6,4))
ax = sns.barplot(x='visit_comparison_group', y='actual_conversion', data=group_means, palette='Blues_d')
ax.set_title("Actual Conversion Rate by Visit Comparison Group")
ax.set_xlabel("Actual vs. Assigned Visits")
ax.set_ylabel("Mean Actual Conversion Rate")
# Optionally annotate exact values on top of each bar
for i, row in group_means.iterrows():
    ax.text(i, row['actual_conversion'] + 0.01, f"{row['actual_conversion']:.2f}", 
            ha='center', va='bottom', color='black')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
df_customers['delta_opt_minus_actual'] = df_customers['optimal_expected_conversion'] - df_customers['actual_conversion']

delta_group = df_customers.groupby('visit_comparison_group')['delta_opt_minus_actual'].mean().reset_index()

plt.figure(figsize=(6,4))
ax = sns.barplot(x='visit_comparison_group', y='delta_opt_minus_actual', data=delta_group, palette='Reds_d')
ax.set_title("Gap: Model's Optimal Predicted Conversion - Actual Conversion")
ax.set_xlabel("Actual vs. Assigned Visits")
ax.set_ylabel("Mean Difference (Probability - Actual)")

for i, row in delta_group.iterrows():
    ax.text(i, row['delta_opt_minus_actual'] + 0.001, f"{row['delta_opt_minus_actual']:.2f}", 
            ha='center', va='bottom', color='black')
plt.axhline(0, color='black', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# We'll compute group means of the predicted prob under actual visits 
# and the predicted prob under optimal visits.
comparison_df = df_customers.groupby('visit_comparison_group').agg({
    'optimal_expected_conversion': 'mean',
    'actual_conversion_prob': 'mean'
}).reset_index()

plt.figure(figsize=(8,4))
melted = comparison_df.melt(id_vars='visit_comparison_group', 
                            value_vars=['optimal_expected_conversion', 'actual_conversion_prob'], 
                            var_name='Conversion Type', value_name='Mean Prob')

ax = sns.barplot(x='visit_comparison_group', y='Mean Prob', hue='Conversion Type', 
                 data=melted, palette='Set2')
ax.set_title("Optimal vs. Actual Conversion Probability by Group")
ax.set_xlabel("Actual vs. Assigned Visits")
ax.set_ylabel("Mean Conversion Probability")
ax.set_ylim(0, 1)
plt.legend(title='')
plt.tight_layout()
plt.show()
```
